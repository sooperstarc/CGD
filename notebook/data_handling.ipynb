```python
root_dir = 'data/custom'
```

```python
# 폴더 내에 있는 이미지들을 '폴더명_00.jpg' 형태로 변경하는 코드

for (root, dirs, files) in os.walk(root_dir):
        # print("# root : " + root)
        # if len(dirs) > 0:
        #     for dir_name in dirs:
        #         print("dir: " + dir_name)

        # if len(files) > 0:
        #     for file_name in files:
        #         # print(dirs)
        #         print("file: " + file_name)

        if len(files) > 0:
            dir_name = os.path.basename(root)
            for index, file_name in enumerate(files):
                print("file: " + file_name)
                ext = os.path.splitext(file_name)[1]
                new_name = os.path.join(root, f'{dir_name}_{str(index).zfill(2)}{ext}')
                # print(new_name)
                os.rename(os.path.join(root, file_name), new_name)
```
